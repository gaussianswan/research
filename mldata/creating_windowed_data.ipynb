```python
import pandas as pd

from data import OHLCVData
```

```python
ticker = 'META'

meta = OHLCVData.from_yahoo_finance(ticker = ticker, name = ticker)
```

```python
close_prices = meta.get_series(column_name='Close')
```

```python
windowed_data = pd.DataFrame()

window_length = 10

for i in range(1, window_length+1): 

    prev = close_prices.shift(i)
    windowed_data[f'Prev_{i}'] = prev
```
