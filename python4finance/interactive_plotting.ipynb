```python
import pandas as pd
import cufflinks as cf 
import plotly.offline as plyo 

plyo.init_notebook_mode(connected = True)
```

```python
coin_data = pd.read_csv("pricing_data/BTC_daily_data.csv")
coin_data.set_index('TIME', inplace=True)
```

```python
plyo.iplot(
    coin_data.iplot(
        asFigure = True, 
        title = 'BTC price information', 
        xTitle = 'Date and Time', 
        yTitle = 'Value', 
        size = 3.5, 
        scale = 'log'
    ), 
    image = 'png', 
    filename = 'bitcoin_chart'
)
```

```python
qf = cf.QuantFig(
    coin_data.loc['2020-01-01':], 
    title = 'BTC-USD candles', 
    legend = 'top', 
    name = 'BTC/USD'
)

# You can add some technical indicators to the chart here
qf.add_bollinger_bands(periods = 15, boll_std=2)
qf.add_rsi(periods = 14, showbands = False)

# you can add a support line here for quick plotting
qf.add_support(date = '2021-01-01', on = 'close', mode = 'toend')
plyo.iplot(
    qf.iplot(asFigure = True), 
    image = 'png', 
    filename = 'BTC_candles'
)
```
